```python
# Sample text
text = "I love learning natural language processing with Python"

# Tokenize text (split by spaces)
tokens = text.split()

# Create bigrams (pairs of consecutive words)
bigrams = [(tokens[i], tokens[i+1]) for i in range(len(tokens)-1)]

# Display bigrams
print("Bigrams:", bigrams)
```
