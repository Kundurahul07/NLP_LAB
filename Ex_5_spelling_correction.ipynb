```python
from textblob import TextBlob

# Input sentence with mistakes
text = "I havv a speling eror in this sentense."

# Create a TextBlob object
blob = TextBlob(text)

# Correct the spelling
corrected = blob.correct()

print("Original:", text)
print("Corrected:", corrected)
```
