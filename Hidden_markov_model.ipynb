```python
from collections import Counter

states = ('Noun','Verb')
obs = ('time','flies','like','an','arrow')

start_p = {'Noun':0.6,'Verb':0.4}
trans_p = {'Noun':{'Noun':0.1,'Verb':0.9}, 'Verb':{'Noun':0.8,'Verb':0.2}}
emit_p = {'Noun':{'time':0.5,'flies':0.3,'like':0.1,'arrow':0.1},
          'Verb':{'time':0.1,'flies':0.4,'like':0.4,'arrow':0.1}}

def viterbi(obs, states, start_p, trans_p, emit_p):
    V = [{st: start_p[st]*emit_p[st].get(obs[0],0) for st in states}]
    for t in range(1,len(obs)):
        V.append({st: max(V[t-1][s]*trans_p[s][st]*emit_p[st].get(obs[t],0) for s in states) for st in states})
    return [max(V[-1], key=V[-1].get)]

print("Predicted POS Tag:", viterbi(obs,states,start_p,trans_p,emit_p))
```
