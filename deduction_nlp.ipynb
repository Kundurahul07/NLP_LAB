```python
facts = {
    "Socrates": "human",
    "Plato": "human"
}

rules = {
    "human": "mortal"
}

def deduce(entity):
    if entity in facts:
        fact = facts[entity]
        if fact in rules:
            return f"{entity} is {rules[fact]}"
    return f"No deduction for {entity}"

print(deduce("Socrates"))
print(deduce("Plato"))
print(deduce("Zeus"))
```
