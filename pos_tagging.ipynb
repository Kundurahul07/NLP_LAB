```python
import spacy

# Load English language model
nlp = spacy.load("en_core_web_sm")

# Example text
text = "ChatGPT generates useful code snippets."

# Process the text
doc = nlp(text)

# Print tokens and their part-of-speech tags
for token in doc:
    print(f"{token.text:<15} {token.pos_}")

```
