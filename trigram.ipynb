```python
# Sample text
text = "I love learning natural language processing with Python"

# Tokenize text (split by spaces)
tokens = text.split()

# Create trigrams (triplets of consecutive words)
trigrams = [(tokens[i], tokens[i+1], tokens[i+2]) for i in range(len(tokens)-2)]

# Display trigrams
print("Trigrams:", trigrams)
```
